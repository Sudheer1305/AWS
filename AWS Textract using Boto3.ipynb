```python
import boto3
```

```python
client = boto3.client('textract')
```

```python
bucket = 'aws-bucket-sudheer'
document = 'aadhar.jpg'
```

```python
response = client.detect_document_text(
    Document= {
        'S3Object':{
            'Bucket': bucket,
            'Name': document
        }
    })
```

```python
print(response)
```

```python
text = []
```

```python
for item in response['Blocks']:
    if item['BlockType'] == 'LINE':
        if item['Confidence'] >= 99:
            #print (item['Text'])
            text.append(item['Text'])
```

```python
Name = text[1]
address = ''
Phone = text[9]
```

```python
for t in text[2:8]:
    address += t + ', '
```

```python
print('Name: ',Name)
print('Address: ', address)
print('Phone: ', Phone)
```
